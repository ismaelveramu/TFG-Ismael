```python
import pandas as pd
```

```python
#Llegeix taula
table = pd.read_excel('GRINDB_2023-04-18.xlsx')
print(table)
```

```python
#Eliminem columnes que no ens interessen
#Ens qudem només amb: Gene, Amino acid position, Gain/Loss, Disease-association, Mutation Category 
taula2=table.drop(['Unnamed: 0', 'Database','Reference Amino acid','Mutated Amino acid','Equivalent position','Aminoacid change','Nucleotide change','Domain','Functional Information','Phenotype category','Clinical description','NM_code','NP_code','Pubmed','Code','Ensembl','VariationID','AlleleID(s)','GRCh37Chromosome','GRCh37Location','GRCh38Chromosome','GRCh38Location'], axis=1)
print(taula2)
```

```python
#Per GRIN1
#Agafem de la taula només GRIN1 i creem nova taula
G1=taula2.loc[taula2['Gene']=='GRIN1']
#De la nova taula agafem només Missense i creem una altra nova taula
G1_1=G1.loc[G1['Mutation Category']=='Missense']
G1_1
```

```python
#GRIN 1 LLISTA POSICIÓ AA
#creo una llista amb la posició del aminoàcid
llista1=list(G1_1['Amino acid position'])
#la ordeno per ordre creixent
llista1.sort()
#creo llista buida
L1=[]
#inicialitzo comptador a 0
comp=0
#mentres que la llargada de la llista sigui diferent al comptador
while len(llista1)!=comp:
    #poso que "h" és resi, la posició de la llista de aa, i or
    h=['resi',llista1[comp],'or ']
    #afegeixo h a la llista buida L1
    L1.append(h)
    #sumo 1 al comptador per canviar de posició
    comp=comp+1
#paso la llista a cadena
s = ','.join(map(str, L1))
#elimino: [ ] , '
ss=s.replace('[','')
sss=ss.replace(',','')
ssss=sss.replace(']','')
sssss=ssss.replace("'","")
#ara borrem el "or" final 
#faig una funció per borrar el or final
def remove_last_or(chain):
    #Divideix la cadena en diferents elements separats per "or"
    elements = chain.split("or")
    #Per si nomes hagues un a la llista, retorna la cadena tal com està
    if len(elements) == 1:
        return chain
    #Treu qualsevol espai en blanc inicial o final de cada element
    cleaned_elements = [elem.strip() for elem in elements]
    #Elimina l'últim element, que hauria de ser una cadena buida després de l'últim "or"
    cleaned_elements.pop()
    #Torna a unir els elements "netejats" fent servir "or" com a separador
    new_chain = " or ".join(cleaned_elements)
    return new_chain
Grin1_aa = remove_last_or(sssss)
Grin1_aa
```

```python
G2A=taula2.loc[taula2['Gene']=='GRIN2A']
G2A_1=G2A.loc[G2A['Mutation Category']=='Missense']
G2A_1
```

```python
#GRIN 2A LLISTA POSICIÓ AA
#creo una llista amb la posició del aminoàcid
llista2A=list(G2A_1['Amino acid position'])
#la ordeno per ordre creixent
llista2A.sort()
#creo llista buida
L2A=[]
#inicialitzo comptador a 0
comp=0
#mentres que la llargada de la llista sigui diferent al comptador
while len(llista2A)!=comp:
    #poso que "h" és resi, la posició de la llista de aa, i or
    h=['resi',llista2A[comp],'or ']
    #afegeixo h a la llista buida L1
    L2A.append(h)
    #sumo 1 al comptador per canviar de posició
    comp=comp+1
#paso la llista a cadena
s = ','.join(map(str, L2A))
#elimino: [ ] , '
ss=s.replace('[','')
sss=ss.replace(',','')
ssss=sss.replace(']','')
sssss=ssss.replace("'","")
#ara borrem el "or" final 
Grin2A_aa = remove_last_or(sssss)
Grin2A_aa
```

```python
G2B=taula2.loc[taula2['Gene']=='GRIN2B']
G2B_1=G2B.loc[G2B['Mutation Category']=='Missense']
G2B_1
```

```python
#GRIN 2B LLISTA POSICIÓ AA
#creo una llista amb la posició del aminoàcid
llista2B=list(G2B_1['Amino acid position'])
#la ordeno per ordre creixent
llista2B.sort()
#creo llista buida
L2B=[]
#inicialitzo comptador a 0
comp=0
#mentres que la llargada de la llista sigui diferent al comptador
while len(llista2B)!=comp:
    #poso que "h" és resi, la posició de la llista de aa, i or
    h=['resi',llista2B[comp],'or ']
    #afegeixo h a la llista buida L1
    L2B.append(h)
    #sumo 1 al comptador per canviar de posició
    comp=comp+1
#paso la llista a cadena
s = ','.join(map(str, L2B))
#elimino: [ ] , '
ss=s.replace('[','')
sss=ss.replace(',','')
ssss=sss.replace(']','')
sssss=ssss.replace("'","")
#ara borrem el "or" final 
Grin2B_aa = remove_last_or(sssss)
Grin2B_aa
```
